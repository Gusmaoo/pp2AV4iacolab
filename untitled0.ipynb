```python
import os

# LOCAL CORRETO PARA ESCRITA
base_dir = "/content/images_classified"

classes = ["present", "absent", "distracted", "sleeping", "empty"]

for cls in classes:
    os.makedirs(os.path.join(base_dir, cls), exist_ok=True)

print("Pastas criadas com sucesso em:", base_dir)
```

```python
import shutil
import random

original_images = f"{path}/dataset/images"
all_images = os.listdir(original_images)

for img in all_images:
    chosen_class = random.choice(classes)
    shutil.copy(
        os.path.join(original_images, img),
        os.path.join(base_dir, chosen_class, img)
    )
```

```python
from torchvision import datasets, transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

dataset_path = base_dir

dataset = datasets.ImageFolder(
    root=dataset_path,
    transform=transform
)

print("Classes encontradas:", dataset.classes)
```

```python
from torch.utils.data import DataLoader, random_split

train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size

train_dataset, test_dataset = random_split(
    dataset, [train_size, test_size]
)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print("Treino:", len(train_dataset))
print("Teste:", len(test_dataset))
```

```python
import torch
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Usando:", device)

model = models.resnet18(pretrained=True)

model.fc = nn.Linear(model.fc.in_features, len(dataset.classes))

model = model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total
    print(f"poca [{epoch+1}/{num_epochs}] - Loss: {running_loss:.4f} - Acurcia: {accuracy:.2f}%")
```

```python
import os
import random
import streamlit as st
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image
import mysql.connector

st.set_page_config(page_title="Classroom Monitoring App", layout="centered")

CLASSES = ["present", "absent", "distracted", "sleeping", "empty"]

POSSIBLE_IMAGE_DIRS = [
    "dataset/images",
    "/content/dataset/images",
    "/kaggle/input/classroom-monitoring-dataset/dataset/images"
]

IMAGE_DIR = None
for p in POSSIBLE_IMAGE_DIRS:
    if os.path.exists(p):
        IMAGE_DIR = p
        break

if IMAGE_DIR is None:
    st.error("Pasta de imagens no encontrada")
    st.stop()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = models.resnet18(weights=None)
model.fc = nn.Linear(model.fc.in_features, len(CLASSES))
model.load_state_dict(torch.load("model/resnet18_classroom.pth", map_location=device))
model.eval()
model.to(device)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

def salvar_predicao(classe, confianca):
    conn = mysql.connector.connect(
        host="centerbeam.proxy.rlwy.net",
        port=20482,
        user="root",
        password="NSlyMvgnzErvxMQCNdwIEqEyezlfWaCQ",
        database="railway",
        connection_timeout=10
    )
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO predictions (predicted_class, confidence) VALUES (%s, %s)",
        (classe, confianca)
    )
    conn.commit()
    cursor.close()
    conn.close()

def carregar_historico():
    conn = mysql.connector.connect(
        host="centerbeam.proxy.rlwy.net",
        port=20482,
        user="root",
        password="NSlyMvgnzErvxMQCNdwIEqEyezlfWaCQ",
        database="railway",
        connection_timeout=10
    )
    cursor = conn.cursor()
    cursor.execute("""
        SELECT predicted_class, confidence, created_at
        FROM predictions
        ORDER BY created_at DESC
        LIMIT 10
    """)
    dados = cursor.fetchall()
    cursor.close()
    conn.close()
    return dados

st.title(" Classroom Monitoring App")
st.write("Aplicao de classificao automtica de comportamento em sala de aula")
st.success("App carregado com sucesso")

imagens = [f for f in os.listdir(IMAGE_DIR) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

if not imagens:
    st.error("Nenhuma imagem encontrada")
    st.stop()

imagem_escolhida = random.choice(imagens)
imagem_path = os.path.join(IMAGE_DIR, imagem_escolhida)

image = Image.open(imagem_path).convert("RGB")
st.image(image, caption=f"Imagem selecionada: {imagem_escolhida}", width="stretch")

img_tensor = transform(image).unsqueeze(0).to(device)

with torch.no_grad():
    outputs = model(img_tensor)
    probs = torch.softmax(outputs, dim=1)
    confidence, predicted = torch.max(probs, 1)

predicted_class = CLASSES[predicted.item()]
confidence_value = float(confidence.item())

salvar_predicao(predicted_class, confidence_value)

st.subheader("Resultado da predio")
st.success(f"Classe prevista: {predicted_class}")
st.info(f"Confiana: {confidence_value:.2f}")

st.subheader("ltimas predies")
historico = carregar_historico()
for classe, conf, data in historico:
    st.write(f"{data}  {classe} ({conf:.2f})")
```
